# Credit_Card_Default_Prediction_using_Machine_Learning

## Problem Statement
The goal of this project is to predict whether a customer will default on a loan or not using machine learning. This helps banks reduce financial risk.

## Dataset Description
Features used:
- Income
- Age
- Loan
- Loan to Income

Target:
- Default (0 = No, 1 = Yes)

## Model Used
Logistic Regression is used because it is suitable for binary classification problems and is simple and effective.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Credit%20Default.csv')
df.head()

df.info()

y = df['Default']
X = df.drop('Default', axis=1)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2529)

X_train.shape, X_test.shape

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))

import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

## Results
- Accuracy: 93%
- Recall (Defaulters): 94%

The model performs well and is able to correctly identify most defaulters.

## Conclusion
The model successfully predicts whether a customer will default or not. Recall is prioritized to avoid missing defaulters, which can lead to financial loss.
